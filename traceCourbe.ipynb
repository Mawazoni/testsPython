```python
import numpy as np
import matplotlib.pyplot as plt

def maFonction (x):
  y = 2*x**2 + 3*x + 4
  return y

x = np.arange(-500,500,1)
y = []
for e in x:
  y.append(maFonction(e))
y = np.array(y)
plt.plot(x,y)
```
