```python
import numpy as np
import matplotlib.pyplot as plt
import math

def f(x):
  return np.log(1+x)+(0.5*x**2)-(x)

x = np.arange(0,1000000,1)

y = []
for e in x:
  y.append(f(e))
y = np.array(y)
plt.plot(x,y)

```
